```python
# Comprehensive provider analysis with NPI, TIN, and POS information
import polars as pl
import re

# Helper function to slugify payer names (from Data_Schema.md)
def slugify(s: str) -> str:
    return re.sub(r'[^a-z0-9]+', '_', s.lower()).strip('_') if s else None

# Read all the necessary tables
fact = pl.read_parquet("core/data/gold/fact_rate.parquet")
dim_code = pl.read_parquet("core/data/silver/dim_code/dim_code.parquet")
xref_group_npi = pl.read_parquet("core/data/silver/xref_group_npi/xref_group_npi.parquet")
xref_group_tin = pl.read_parquet("core/data/silver/xref_group_tin/xref_group_tin.parquet")
xref_pos = pl.read_parquet("core/data/silver/xref_pos/xref_pos.parquet")
dim_pos_set = pl.read_parquet("core/data/silver/dim_pos_set/dim_pos_set.parquet")

# Filter fact for specific criteria (adjust as needed)
fact_filtered = fact.filter(
    (pl.col("state") == "GA") & 
    (pl.col("year_month") == "2025-08") &
    (pl.col("code_type") == "CPT") &
    (pl.col("code") == "10121")  # Example: CPT 10121
)

# Add payer_slug to fact for joins (computed on the fly)
fact_with_slug = fact_filtered.with_columns(
    pl.col("reporting_entity_name").map_elements(slugify).alias("payer_slug")
)

# Join with code descriptions
fact_with_codes = fact_with_slug.join(
    dim_code, 
    on=["code_type", "code"], 
    how="left"
)

# Join with NPI information
fact_with_npis = fact_with_codes.join(
    xref_group_npi,
    on=["year_month", "payer_slug", "pg_uid"],
    how="left"
)

# Join with TIN information
fact_with_tins = fact_with_npis.join(
    xref_group_tin,
    on=["year_month", "payer_slug", "pg_uid"],
    how="left"
)

# Join with POS information to get the actual POS codes
fact_with_pos = fact_with_tins.join(
    xref_pos,
    on="pos_set_id",
    how="left"
)

# Join with POS set descriptions to get the full list of POS codes
fact_with_pos_details = fact_with_pos.join(
    dim_pos_set,
    on="pos_set_id",
    how="left"
)

# Select and organize the final columns
comprehensive_result = fact_with_pos_details.select([
    # Basic fact information
    "fact_uid",
    "year_month",
    "state",
    "reporting_entity_name",
    "billing_class",
    
    # Code information
    "code_type",
    "code", 
    "code_desc",
    "name",
    
    # Rate information
    "negotiated_rate",
    "negotiated_type",
    "negotiation_arrangement",
    "expiration_date",
    
    # Provider information
    "pg_uid",
    "provider_group_id_raw",
    "npi",
    "tin_type",
    "tin_value",
    
    # POS information
    "pos_set_id",
    "pos",
    "service_codes",
    
    # Additional context
    "payer_slug"
]).sort(["reporting_entity_name", "negotiated_rate"])

print("Comprehensive provider analysis with NPI, TIN, and POS information:")
print(f"Total records: {comprehensive_result.height}")
print("\nSample results:")
print(comprehensive_result.head(10))

# Summary statistics
print("\n" + "="*80)
print("SUMMARY STATISTICS")
print("="*80)

# Rate statistics by provider type
rate_stats = comprehensive_result.group_by(["tin_type"]).agg([
    pl.len().alias("count"),
    pl.col("negotiated_rate").mean().alias("avg_rate"),
    pl.col("negotiated_rate").min().alias("min_rate"),
    pl.col("negotiated_rate").max().alias("max_rate"),
    pl.col("negotiated_rate").quantile(0.5, "nearest").alias("median_rate")
]).sort("count", descending=True)

print("\nRate statistics by TIN type:")
print(rate_stats)

# POS distribution
pos_dist = comprehensive_result.group_by("pos").agg([
    pl.len().alias("count"),
    pl.col("negotiated_rate").mean().alias("avg_rate")
]).sort("count", descending=True)

print("\nPOS code distribution:")
print(pos_dist.head(10))

# Provider count by NPI
provider_count = comprehensive_result.filter(
    pl.col("npi").is_not_null()
).group_by("npi").agg([
    pl.len().alias("rate_count"),
    pl.col("negotiated_rate").mean().alias("avg_rate")
]).sort("rate_count", descending=True)

print(f"\nTop 10 providers by number of rates:")
print(provider_count.head(10))
```

### Develop Data Dictionary

```python
# Comprehensive Data Dictionary Generator
import polars as pl
import duckdb
from pathlib import Path

# Define all the core data files
core_files = {
    "xref_pg_member_tin": "core/data/xrefs/xref_pg_member_tin.parquet",
    "xref_pg_member_npi": "core/data/xrefs/xref_pg_member_npi.parquet",
    "dim_provider_group": "core/data/dims/dim_provider_group.parquet",
    "dim_pos_set": "core/data/dims/dim_pos_set.parquet",
    "dim_payer": "core/data/dims/dim_payer.parquet",
    "dim_npi": "core/data/dims/dim_npi.parquet",
    "dim_npi_address": "core/data/dims/dim_npi_address.parquet",
    "dim_code": "core/data/dims/dim_code.parquet",
    "dim_code_cat": "core/data/dims/dim_code_cat.parquet",
    "fact_rate": "core/data/gold/fact_rate.parquet"
}

def analyze_table(table_name: str, file_path: str):
    """Analyze a single table and return metadata"""
    if not Path(file_path).exists():
        return {"error": f"File not found: {file_path}"}
    
    try:
        # Read the table
        df = pl.read_parquet(file_path)
        
        # Basic info
        row_count = df.height
        col_count = len(df.columns)
        
        # Column analysis
        columns = {}
        for col in df.columns:
            col_info = {
                "dtype": str(df[col].dtype),
                "null_count": df[col].null_count(),
                "null_pct": round((df[col].null_count() / row_count) * 100, 2) if row_count > 0 else 0,
                "unique_count": df[col].n_unique(),
                "sample_values": df[col].drop_nulls().head(3).to_list() if df[col].null_count() < row_count else []
            }
            columns[col] = col_info
        
        # Sample data
        sample_data = df.head(3).to_dicts() if row_count > 0 else []
        
        return {
            "row_count": row_count,
            "col_count": col_count,
            "columns": columns,
            "sample_data": sample_data
        }
    
    except Exception as e:
        return {"error": str(e)}

# Analyze all tables
print("=== COMPREHENSIVE DATA DICTIONARY ===\n")

table_metadata = {}
for table_name, file_path in core_files.items():
    print(f"📊 Analyzing {table_name}...")
    metadata = analyze_table(table_name, file_path)
    table_metadata[table_name] = metadata
    
    if "error" in metadata:
        print(f"❌ {metadata['error']}\n")
        continue
    
    print(f"   Rows: {metadata['row_count']:,}")
    print(f"   Columns: {metadata['col_count']}")
    print(f"   Columns: {list(metadata['columns'].keys())}")
    print()

# Generate detailed data dictionary
print("=" * 80)
print("DETAILED DATA DICTIONARY")
print("=" * 80)

for table_name, metadata in table_metadata.items():
    if "error" in metadata:
        continue
        
    print(f"\n## {table_name.upper()}")
    print(f"**File:** `{core_files[table_name]}`")
    print(f"**Rows:** {metadata['row_count']:,}")
    print(f"**Columns:** {metadata['col_count']}")
    
    print("\n### Column Details:")
    for col_name, col_info in metadata['columns'].items():
        print(f"- **{col_name}** ({col_info['dtype']})")
        print(f"  - Nulls: {col_info['null_count']:,} ({col_info['null_pct']}%)")
        print(f"  - Unique values: {col_info['unique_count']:,}")
        if col_info['sample_values']:
            print(f"  - Sample: {col_info['sample_values']}")
        print()
    
    if metadata['sample_data']:
        print("### Sample Data:")
        sample_df = pl.DataFrame(metadata['sample_data'])
        print(sample_df)
        print()

# Generate relationship analysis
print("=" * 80)
print("RELATIONSHIP ANALYSIS")
print("=" * 80)

# Check foreign key relationships
if "fact_rate" in table_metadata and "dim_payer" in table_metadata:
    print("\n### Fact Table Relationships:")
    
    # Load fact table for relationship checks
    fact_df = pl.read_parquet("core/data/gold/fact_rate.parquet")
    
    # Check each dimension relationship
    relationships = {
        "payer_slug": "dim_payer",
        "code_type,code": "dim_code", 
        "pg_uid": "dim_provider_group",
        "pos_set_id": "dim_pos_set"
    }
    
    for fk_cols, dim_table in relationships.items():
        if dim_table in table_metadata:
            print(f"\n**Fact.{fk_cols} -> {dim_table}:**")
            
            # Check if all fact keys exist in dimension
            if "," in fk_cols:
                # Composite key
                fact_keys = fact_df.select(fk_cols.split(",")).unique()
                dim_keys = pl.read_parquet(f"core/data/dims/{dim_table}.parquet").select(fk_cols.split(",")).unique()
                missing = fact_keys.join(dim_keys, on=fk_cols.split(","), how="anti")
            else:
                # Single key
                fact_keys = fact_df.select(fk_cols).unique()
                dim_keys = pl.read_parquet(f"core/data/dims/{dim_table}.parquet").select(fk_cols).unique()
                missing = fact_keys.join(dim_keys, on=fk_cols, how="anti")
            
            print(f"  - Fact keys: {fact_keys.height:,}")
            print(f"  - Dim keys: {dim_keys.height:,}")
            print(f"  - Missing: {missing.height:,}")
            print(f"  - ✅ Valid" if missing.height == 0 else f"  - ❌ {missing.height} orphaned keys")

print("\n" + "=" * 80)
print("DATA DICTIONARY COMPLETE")
print("=" * 80)
```
